```python
melt_sheet1['SKU'] = 'Total US'
melt_sheet1.head()
```

```python
concat_df = pd.DataFrame()

for i in range(1,66):
    sheet = pd.read_excel('Tables/2. 2016-2020 Consumption by week_USC CONFIDENTIAL.xlsx', sheet_name = i, header = 8).\
                    rename(columns={'Unnamed: 0': 'Product'})
    melt_sheet = pd.melt(sheet, 
                      id_vars = ['Product'], 
                      var_name = 'Week Ending',
                      value_name = 'Consumption')
    melt_sheet['SKU'] = summary.loc[i,'Unnamed: 3']
    concat_df = pd.concat([concat_df, melt_sheet], axis = 0, ignore_index = True)
```

```python
concat_df.shape
```

```python
concat_df = concat_df.fillna(0)
concat_df['Week Ending'] = concat_df['Week Ending'].str.extract('(\d+-\d+-\d+)')
concat_df.head()
```

```python
concat_df['Week Ending'] = pd.to_datetime(concat_df['Week Ending'])
concat_df.head()
```

```python
concat_df['SKU'] = concat_df['SKU'].str.replace(' - Multi Outlet', '')
concat_df.head()
```

```python
concat_df.info()
```

```python
concat_df['SKU'].unique() 
```

```python
US_consumption = concat_df[concat_df['SKU']=='Total US']

plt.figure(figsize = (20,10))

sns.lineplot(data = US_consumption,
             x = 'Week Ending',
             y = 'Consumption',
             hue = 'Product')
```

```python
plt.figure(figsize = (20,10))
US_consumption.groupby('Product')['Consumption'].sum().plot(kind = 'bar')
```
