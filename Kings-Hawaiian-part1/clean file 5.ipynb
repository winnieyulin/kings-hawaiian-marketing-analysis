```python
concat_df = pd.DataFrame()

for i in range(1,66):
    sheet = pd.read_excel('Tables/5. 2016-2020 ACV by Display Feature by week_USC CONFIDENTIAL.xlsx', sheet_name = i, header = 8).\
                    rename(columns={'Unnamed: 0': 'Product'})
    melt_sheet = pd.melt(sheet, 
                      id_vars = ['Product'], 
                      var_name = 'Week Ending',
                      value_name = 'ACV')
    melt_sheet['Geography'] = summary.loc[i,'geography']
    concat_df = pd.concat([concat_df, melt_sheet], axis = 0, ignore_index = True)
```

```python
concat_df = concat_df.fillna(0)
concat_df['Week Ending'] = concat_df['Week Ending'].str.extract('(\d+-\d+-\d+)')
concat_df.head()
```

```python
concat_df['Week Ending'] = pd.to_datetime(concat_df['Week Ending'])
concat_df.head()
```

```python
concat_df['Geography'] = concat_df['Geography'].str.replace(' - Multi Outlet', '')
concat_df.head()
```

```python
US_ACV = concat_df[concat_df['Geography']=='Total US']

plt.figure(figsize = (20,10))

sns.lineplot(data = US_ACV,
             x = 'Week Ending',
             y = 'ACV',
             hue = 'Product')
```
