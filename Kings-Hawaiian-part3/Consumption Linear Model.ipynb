```python
import pandas as pd
import statsmodels.api as sm
```

```python
consumption = pd.read_csv('Tables/consumption adjusted.csv', index_col=0)
consumption
```

```python
y = consumption.Consumption
```

```python
y
```

```python
X = consumption.iloc[:,5:-2]
```

```python
X
```

```python
display = pd.read_csv('Tables/Display.csv')
```

```python
display = display.iloc[:,4:].reset_index(drop =True)
```

```python
display
```

```python
X = pd.concat([X, display], axis=1)
X = sm.add_constant(X)
```

```python
X
```

```python
model = sm.OLS(y, X)
result = model.fit()
print(result.summary())
```
