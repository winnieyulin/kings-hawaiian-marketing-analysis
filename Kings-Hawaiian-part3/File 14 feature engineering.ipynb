```python
import pandas as pd
```

```python
# Load data
data = pd.read_csv('Tables/14. Ongoing_promos_placement_by_retailer Cleaned.csv')
data = data.iloc[:,1:].reset_index(drop = True)
data
```

```python
data.columns
```

```python
# Remove trivial info
data = data[['Geography', 'Product',
       'Chicken Program/Meal Deal', 'Secondary Deli', 'Secondary Bakery',
       'Secondary Center store', 'Retail Coverage', 'Volume Sales',
       'Volume Sales Any Merch','Volume Sales Any Price Reduction',
       'Volume Sales Any Feature', 'Volume Sales Any Display','Holiday_Back to School ', 'Holiday_Christmas', 'Holiday_Christmas/NY',
       'Holiday_Easter Week ', 'Holiday_Easter Weeks', 'Holiday_Halloween ',
       'Holiday_July 4th ', 'Holiday_Labor Day ', 'Holiday_Memorial Week ',
       'Holiday_SuperBowl ', 'Holiday_Thanksgiving ']]
data
```

```python
data.Geography.value_counts()
```

```python
# Map Geography into corresponding category
data.loc[(data["Geography"] != 'Target Corp')&(data["Geography"] != 'Walmart Corp'),'Geography'] = 'Food'
data.loc[data["Geography"] == 'Walmart Corp', 'Geography'] = 'Mass'
data.loc[data["Geography"] == 'Target Corp','Geography'] = 'Mass'
```

```python
data.Geography.value_counts()
```

```python
# tranform Geography into dummy variables
dummy = pd.get_dummies(data.Geography, drop_first=True)
data['Geography'] = dummy['Mass']
data
```

```python
# Check abnormal values
for i in ['Chicken Program/Meal Deal', 'Secondary Deli',
          'Secondary Bakery', 'Secondary Center store', 'Retail Coverage']:
    print(data[i].value_counts())
```

```python
# remove abnormal values
for i in ['Chicken Program/Meal Deal', 'Secondary Deli',
          'Secondary Bakery', 'Secondary Center store']:
    data.loc[(data[i] != 0)&(data[i] != 1), i] = 1
for i in ['Chicken Program/Meal Deal', 'Secondary Deli',
          'Secondary Bakery', 'Secondary Center store', 'Retail Coverage']:
    print(data[i].value_counts())
```

```python
data.columns
```

```python
# obtain products' names
products = []
for i in data['Product'].value_counts().index:
    products.append(i)
products
```

```python
# create separate dataset by products & adjust orders of columns
for i in products:
    product = data.loc[data['Product'] == i, ['Volume Sales', 'Volume Sales Any Merch', 
        'Volume Sales Any Price Reduction',
       'Volume Sales Any Feature', 'Volume Sales Any Display',
    'Geography', 'Chicken Program/Meal Deal', 'Secondary Deli',
       'Secondary Bakery', 'Secondary Center store', 'Retail Coverage',
         'Holiday_Back to School ',
       'Holiday_Christmas', 'Holiday_Christmas/NY', 'Holiday_Easter Week ',
       'Holiday_Easter Weeks', 'Holiday_Halloween ', 'Holiday_July 4th ',
       'Holiday_Labor Day ', 'Holiday_Memorial Week ', 'Holiday_SuperBowl ',
       'Holiday_Thanksgiving ']]
    product.to_csv(f'{i}.csv')
```
